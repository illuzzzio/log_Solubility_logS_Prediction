```python
import pandas as pd 
df = pd.read_csv("solubility_data.csv")
df #by calling the variable we can get the data 
```

```python
#Data Seperation#  # This code helsp su to visualise specifically the logS column 


y = df["logS"]
y
```

```python
# now we will get the data without logS column#


x = df.drop("logS", axis=1)  #axis = 1 , means drop the column (means vertical section), axis = 0 means drop the horizontal row , so here we need to dro pthe column of logS
x
```

```python
# DATA SPLITTING USIGN SCIKIT LEARN #


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
#X_train: Training set for features.
# X_test: Testing set for features.
# y_train: Training set for the target variable.
# y_test: Testing set for the target variable.
# test_size: Specifies the proportion of the dataset to be allocated to the test set (e.g., 0.2 for 20%).
# random_state: An integer used to initialize the random number generator, ensuring reproducible splits across multiple runs.

X_train

```

```python
X_test
```

```python
#MODEL-BUILDING(specifically training the model) using LINEAR_REGRESSION_ALGO#

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
LinearRegression()   # Till here the model is ready 
```

```python
#Applyign the model to make a prediction#

y_LR_train_predict = LR.predict(X_train)
y_LR_test_predict  = LR.predict(X_test)

y_LR_train_predict  # prediction made on 80 percent trained data 
```

```python
y_LR_test_predict # these are the predictions made on test data which was 20 percent of the original data
```

```python
#COMPARING AND EVALUATING MODEL-PERFORMANCE#


y_train # y_train was the actual values for the variable y as a fucntion of x i.e logS Values of that 80 percent of data to be trained earlier 
```

```python
y_LR_train_predict # it is the predicted values of y on the 80 percent of data trained earlier 
```

```python
#FINAL MODEL EVALUATION#

from sklearn.metrics import mean_squared_error, r2_score
#mean_squared_error (MSE):

 # Purpose: This function calculates the Mean Squared Error, which is a common metric to quantify the average squared difference between the actual (true) values and the predicted values from a regression model.

 #r2_score (R-squared or Coefficient of Determination):

#Purpose: This function calculates the R-squared score, which measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). In simpler terms, it indicates how well the regression model explains the variability in the target variable


#First lets do it for training models:
LR_train_mse = mean_squared_error(y_train, y_LR_train_predict) 
LR_train_r2 = r2_score(y_train, y_LR_train_predict)
# comparision of both for evaluation of the model 

#Now lets do this for testing models:
LR_test_mse = mean_squared_error(y_test,y_LR_test_predict)
LR_test_r2 = r2_score(y_test,y_LR_test_predict)

print("LR MSE (TRAIN): ",LR_train_mse)
print("LR r2 (TRAIN): ",LR_train_r2)
print("LR MSE (TEST): ",LR_test_mse)
print("LR r2 (TEST): ",LR_test_r2)

# Above i am printing our results 


LR_results = pd.DataFrame(["Linear Regression Model", LR_train_mse, LR_train_r2, LR_test_mse, LR_test_r2]).transpose()
LR_results.columns= ["Method","Training MSE","TrainingR2","Test MSE","Test R2"] # This we are doing to maek data more tidy 
LR_results
```
